```python
codec = cv2.VideoWriter_fourcc(*'DIVX')
vid_writer = cv2.VideoWriter("./output_video.avi", codec, 24, (720, 1280))
```

```python
model = keras.models.load_model("C:/Users/hjhhi/Desktop/Codes/Open_SW/model.h5")
```

```python
import time

green_color=(0, 255, 0)
red_color=(0, 0, 255)

start = time.time()
index=0
while True:
    hasFrame, img_frame = cap.read()
    if not hasFrame:
        print('더 이상 처리할 frame이 없습니다.')
        break
    index += 1
    if index%3 >=1:
        continue
    print('frame :', index, '처리 완료')
#     img_frame = cv2.resize(img_frame, (160, 160), interpolation=cv2.INTER_AREA)
#     img_frame = np.array([img_frame])
#     cv2.rectangle(img_frame, (300, 100, 800, 400), color=green_color, thickness=2)
#     caption = "frame:{}".format(index)
#     pred = model.predict(img_frame)
    
    img_temp = cv2.resize(img_frame, (160, 160), interpolation=cv2.INTER_AREA)
    img_temp = cv2.cvtColor(img_temp, cv2.COLOR_BGR2RGB)
    img_temp = np.reshape(img_temp, (1, 160, 160, 3))
    caption_2 = model.predict(img_temp)
    for i in range(0, len(caption_2[0]),2):
        cv2.line(img_frame, (int(caption_2[0][i]*4.5), int(1280-caption_2[0][i+1]*8)), (int(caption_2[0][i]*4.5), int(1280-caption_2[0][i+1]*8)), red_color, 5)
    cv2.putText(img_frame, str(caption_2), (360, 600), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 1)
    vid_writer.write(img_frame)

print('write 완료 시간:', round(time.time()-start,4))
vid_writer.release()
cap.release()
```
